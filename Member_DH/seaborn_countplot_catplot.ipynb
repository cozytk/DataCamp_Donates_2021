## seaborn.countplot
 - https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style(style='darkgrid')
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
ax = sns.countplot(x='class', data=titanic)
```

```python
ax = sns.countplot(x='class', hue='who', data=titanic)
```

```python
ax = sns.countplot(y='class', hue='who', data=titanic)
```

```python
ax = sns.countplot(x='who', data=titanic, palette='Set3')
```

```python
ax = sns.countplot(x = 'who', 
                   data = titanic,
                   facecolor = (0, 0, 0, 0),
                   linewidth = 5,
                   edgecolor = sns.color_palette('dark', 3)
                  )
```

## seaborn.catplot

```python
g = sns.catplot(x='class', hue='who', col='survived',
                data=titanic, kind='count',
                height=4, aspect=.9)
```

```python
li_x = ['a', 'b', 'c', 'd', 'e']
li_y = [80, 90, 100, 95, 75]
df = pd.DataFrame({'col_x':li_x, 'col_y':li_y})
display(df, sns.catplot(x='col_x', y='col_y', data=df, kind='bar', height=4, aspect=2))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

li_x = ['churn', 'churn', 'churn', 'b', 'b', 'b']
li_y = [80, 90, 100, 95, 75, 85]
li_z = ['f', 'f', 'm', 'f', 'm', 'm']
df = pd.DataFrame({'name':li_x, 'col_y':li_y, 'gender':li_z})
display(df, sns.catplot(x='name', y='col_y', data=df, kind='bar', hue='gender', height=4, aspect=2))
```
