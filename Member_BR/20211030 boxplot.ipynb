#### box plot (상자 그림 이해하기)

*상자 그림 *

= 다섯수치 요약을 보여준다. 그 중에는 최소값, 제1사분위수, 중앙값, 제 3사분위수, 최댓값이 포함된다.

상자는 제 1사분위수에서 제 3 사분위수까지를 나타낸다.
상자를 가로지르는 선은 중앙값이고
나머지 선분은 각 사분위수에서 최솟값과 최댓값까지 이어진다.  
* 두개 이상의 분포를 비교할때 유리하지만  결과 분포에 대한 간단한 요약만 표시되므로  
  다른 데이터 (히스토그램, 산점도 (scatter plot) 와 함께 사용하여  
  데이터를 상세하게 분석한 시각화 자료를 제공할수 있다.

* Q1= 1 사분위수 (데이터의 25% 지점)
* Q2= 2 사분위수 (상자안의 눈금. 즉, 중앙값 (median)
* Q3= 3 사분위수 (데이터의 75% 지점)  

* IQR; inter-quartile range = Q3 - Q1  
  1사분위수와 3사분위수 사이의 간격을 나타내는 사분위수 범위  
* 수염이 끝나는 부분 : 데이터의 최소값과 최대값
* Outer = 1.5 IQR을 벗어난 데이터 (일반적으로 특수값)
* far outs= 혹시라도 outer를 벗어나는 더 멀리에 있는 극단적인 값이 있는경우.  
  (아래그림 max에 해당)
