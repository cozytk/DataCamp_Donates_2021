# Comparison Operators  



## - Pay attention: <= is valid syntax, but =< is not

```python
# Comparison of booleans
print (True == False)
```

```python
# Comparison of strings
print ("pyscript" == "PyScript")
```

```python
# Compare a boolean with an integer
print (True == 1)
```

```python
# 주의: <= is valid syntax, but =< is not
print (True =< 1)
```

```python
# 마찬가지로 >= is valid syntax , but => is not 
True => 1
```

# if, elif, else  

```python
if condition :  # 컨디션 작성후 콜론 (:) 꼭 넣기.
    expression 

elif condition :
    expression

else : # else에는 컨디션이 없지만 : 을 꼭 넣어주어야 코드가 실행된다.
    expression
```

```python
z = 3
if z % 2 == 0 : # 첫 조건인 if를 만족하지 않음으로 다음 조건인 elif 로 넘어감
    print ("z is divisible by 2")
elif z % 3 == 0 :
    print ("z is divisible by 3")
else :
    print ("z is neither divisible by 2 nor by 3 ")
```



  

- 상단처럼 z = 6일때  
 if와 elif를 둘다 만족하는경우에는  
 이미 if가 참이었음으로 아에 elif에 도달하지 않고 if 도달후 바로 출력이 된다.
