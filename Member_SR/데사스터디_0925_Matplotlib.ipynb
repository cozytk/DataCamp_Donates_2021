##Matplotlib 기본

```python
#기본으로 그래프를 만드는 모듈
import matplotlib.pyplot as plt

#리스트 값들을 y값이라고 가정하고, x값을 자동으로 0부터 순서대로 만들어낸다.
plt.plot([1,2,3,4])
plt.show()

#이렇게 plot()함수에 처음 넣은 값은 x,두번째 넣은 값은 y값으로 나타낼 수 있다.
plt.plot([1,2,3,4],[1,4,8,16])
plt.show()
```

###레이블에 있는 데이터 사용

```python
import matplotlib.pyplot as plt
#파이썬 딕셔너리와 같이 레이블에 있는 데이터를 사용할 수 있다.
data_dict = {'data_x':[1,2,3,4,5,6,7], 'data_y':[2,3,5,8,13,16,19]}

plt.plot('data_x','data_y', data=data_dict)
plt.show()
```

###레이블 표시하기 + 디자인적 요소를 곁들여

```python
import matplotlib.pyplot as plt

font1 = {'family': 'serif',
         'color':'b',
         'weight':'bold',
         'size':14
         }
font2= {'family': 'fantasy',
        'color':'deeppink',
        'weight':'normal',
        'size':'xx-large'
        }

#x-y순으로 데이터 입력하기
plt.plot([1,2,3,4],[5,6,7,8])

#labelpad는 여백지정 파라미터, matplotlib 3.3부터는 loc 파라미터로 위치 지정가능('right','top'등등)
plt.xlabel('is x label', labelpad=100,fontdict=font1)
plt.ylabel('is y label', labelpad=50, fontdict=font2)
plt.show()
```

##범례 표시하기

```python
import matplotlib.pyplot as plt
#label을 지정해 plot()함수에 넣고, plt.legend()를 적으면 범례가 표시된다.
plt.plot([1,2,3,4],[2,3,5,10], label='price($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(0.5,0.5))
plt.show()


plt.plot([1,2,3,4],[2,3,5,10], label='price($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='lower right')
plt.show()
```

```python
#열개수 지정하기
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='price($)')
plt.plot([1,2,3,4],[3,4,9,10], label='Demand(#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

#plt.legend(loc='first')
plt.legend(loc='first', ncol=1 ,frameon=True, shadow=True)
plt.show()
```

###Customization


```python
#Axis Labels(축 레이블)
import matplotlib.pyplot as plt
year = [1950,1951,1952,1953]
pop = [2.538,3.57,4.62,5.56]

#데이터 추가하는 방법
year = [1800,1850,1900,2000] + year
pop = [1.0,1.263,1.650,6.7] + pop

plt.plot(year,pop)

#차트 내부 레이블 이름설정방법
plt.xlabel('Year') #차트 x축 레이블 설정
plt.ylabel('Population') #차트 y축 레이블 설정
plt.title('World Population Projection') #차트 이름 설정
plt.yticks([0,4,6],#y축이 0부터 시작할 수 있게 해주는 tick함수와 
           ['0','2B','4B']) #tick의 표시 이름이 있는 목록인 vticks함수에 대한 인수이므로 첫번째 목록과 길이가 같아야 한다. 
plt.show()
```
