###Subset DF(부분집합 데이터프레임)


```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/my_data/서울주택도시공사_전세임대 주택(신혼부부 2형) 공급계획_csv.csv',encoding='CP949',index_col=0)
print(data)

data_apt = pd.DataFrame(data)
print(data_apt.shape)
print(type(data_apt))
```

```python
#값은 같지만 다르게 표현가능하다
print(data_apt["증감(2021-2020)"])
print(data_apt.loc[:,"증감(2021-2020)"])
print(data_apt.iloc[:,2])
```

```python
#불리언 값이 포함된 Series를 도출한다
new_choice = data_apt["증감(2021-2020)"] >  0
new_choice
```

```python
#새로운 변수에 불리언 값이 포함된 시리즈를 넣고, 그 변수를 df에 넣으면, 원하는 값만 도출이 된다!

data_apt[new_choice]
```

```python
#똑같은 값을 도출하지만, 더 간단하게 코드짜기
data_apt[data_apt["증감(2021-2020)"] > 0]
```

###Pandas에서 Numpy속성 활용해 여러조건 결과 도출하기

```python
import numpy as np
data_apt_update = np.logical_and(data_apt["증감(2021-2020)"] > 0 ,data_apt["증감(2021-2020)"] < 3 )
data_apt[data_apt_update]
```
