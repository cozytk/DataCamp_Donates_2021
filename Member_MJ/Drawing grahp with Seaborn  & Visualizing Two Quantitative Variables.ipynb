#### 라이브러리와 데이터 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
```

```python
df =sns.load_dataset('tips')  # seaborn 라이브러리에 내재된 데이터셋 불러오기 
df.head(3)
```

#### Scatterplot  - sns.scatterplot(x , y , data)

```python
plt.rcParams['figure.dpi'] = 80 # dpi 매개 변수의 높은 값을 설정함으로써 고해상도로 그래프를 그릴 수 있습니다.
                                # dpi는 인치당 도트 수를 나타냅니다. 그림에서 인치당 픽셀 수를 나타냅니다.               

fig = plt.figure(facecolor='#FEFBF3') #  set_facecolor(color)는 해당 플롯의 배경색, 즉 얼굴 색을 설정

sns.scatterplot(x = 'total_bill' , y = 'tip' , data =df )
```

#### 매개변수 hue를 통해 값 비교가능 , 자동으로 plot에 범례 추가

```python
plt.rcParams['figure.dpi'] = 80
fig = plt.figure(facecolor='#FEFBF3')

sns.scatterplot(x = 'total_bill' , y = 'tip' , data =df  , hue = 'sex')  
```

#### palette 값 색 변경

```python
plt.rcParams['figure.dpi'] = 80
fig = plt.figure(facecolor='#FEFBF3')


color = {"Male" :"red" , "Female":"Green"}

sns.scatterplot(x = 'total_bill' , y = 'tip' , data =df  , hue = 'sex' ,palette= color )  
```

### Countplot - sns.countplot(x  ,data or y  ,data) 
>범주형 목록을 가져오고 범주당 목록 항목 수를 나타내는 막대를 반환 

```python
plt.rcParams['figure.dpi'] = 80
fig = plt.figure(facecolor='#FEFBF3')

sns.countplot(y = 'time' , data =df , hue ='sex')
```

```python
plt.rcParams['figure.dpi'] = 80
fig = plt.figure(facecolor='#FEFBF3')

sns.countplot(x = 'smoker' , data =df , hue = 'sex' , palette={"Male":"green", "Female":"blue"} , hue_order =['Female','Male'])
```

## replot() 
> 산점도 또는 라인 플롯 를 사용하여 두 양적 변수간의 관계를 확인  
> sns.relplot(x ,y ,data , kind = '' )  
   replot() 사용하면 단일 figure에 subplot을 생성할 수 있다는 이점이 있다.

```python
sns.relplot(data = df , x = "total_bill" , y = "tip" , kind = "scatter" , col = "sex" ,  row = "time" , hue = "smoker")
```
