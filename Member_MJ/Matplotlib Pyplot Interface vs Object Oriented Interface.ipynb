# Matplotlib Pyplot Interface vs Object Oriented Interface

* Matplotlib의 구동 방식, Matplotlib은 두가지 방식의 인터페이스를 제공하고 있습니다.


* Pyplot 인터페이스: Matlab과 같이 커맨드 방식, Matplotlib.pyplot 모듈에 함수로 정의되어 있으며 자동으로 Figure 와 axse 를 생성합니다.


* 객체지향 인터페이스: Matplotlib이 구현된 객체지향라이브러리를 직접 활용하는 방식입니다. 

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# 일반적인 Pyplot Interface 를 이용하는 방식 
x = np.arange(0,10 , 0.2)
y = x
plt.plot(x, y)
plt.title('Pyplot')
plt.xlabel('X')
plt.ylabel('Y')
```

### Object Oriented Interface¶

```python
fig = plt.figure() # 직접 Figure 객체를 생성
ax = fig.add_axes([0,0,1,1]) # 직접 axes를 생성 

#fig.add_axes - 이 함수는 각각의 위치를 지정하는 함수입니다.  
#0.1, 0.1, 0.8, 0.8이라고 작성이 되어있는데 , 첫 0.1은 이미지의 x축의 위치, 두번째 0.1은 이미지의 y축의 시작위치, 세번째 0.8은 이미지의 가로길이, 네번째 0.8은 높이를 의미

```

```python
# 객체지향 Ineterface를 이용하는 방식 
A = np.arange(0,10,0.3)
B = A

fig = plt.figure() # 직접 Figure 객체를 생성
ax = fig.add_axes([0,0,1,1]) # 직접 axes를 생성
ax.plot(A,B) #생성된 axes에 대한 plot 직접호출 
ax.set_title('Object Oriented') 
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
x = np.arange(0,15,0.3)
y = np.arange(0,11,0.2)
y1= y**2


fig, axes = plt.subplots(2,1) # 2개의 ax를 가진 figure 생성
axes[0].plot(x) # 1행 plot 그래프 
axes[0].set_title("It's plot" )
axes[0].set_xlim(0,15)



axes[1].scatter(y,y1) # 2행 scatter 그래프 
axes[1].set_title("It's scatter")

plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95, wspace=0.7, hspace=0.5) #  # subplot 간 간격 조절
```
