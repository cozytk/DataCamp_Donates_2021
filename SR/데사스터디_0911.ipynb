###Object(객체)
숫자,문자,함수와 같은 데이터를 의미한다.
즉,파이썬에서의 모든 데이터는 객체이다.

###Variable(변수)
####객체를 대입할 수 있는 부분이다. '객체를 저장하는 저장소 느낌!'

```python
#여기에서 dog는 "happy"라는 객체를 대입한 변수이다.
dog = "happy" 
print(dog)
```

###Class(클래스)
####와플기계가 클래스라면, 와플은 객체이다. 객체의 행동과 구조를 정의한다.(클래스에 의해 타입이 결정된다)

```python
class num_1:
  pass
```

###Method(메소드)
####클래스 내에 함수를 정의하면 그 함수가 메소드가된다.(객체에 대한 연산을 수행한다),특정 데이터 타입의 객체가 가지고 있는 고유의 함수들을 뜻한다.


###Instance(인스턴스)
####근본적으로는 객체와 같은 개념이지만, 클래스를 통해 만들어진 객체를 뜻한다.

```python
#pie는 클래스를 통해 만들어진 객체이므로 인스턴스라고도 부를 수 있다.
pie = num_1()
print(type(pie))
```
