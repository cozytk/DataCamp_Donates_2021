```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
PATH = os.getenv('HOME') + '/data/'
PATH
```

```python
df = pd.read_csv(PATH + 'EPL Standings 2010-2021.csv')
df
```

```python
df['Team'].value_counts()
```

```python
Teams = ['Manchester United', 'Manchester City', 'Arsenal',
         'Tottenham Hotspur', 'Liverpool', 'Everton',
         'Chelsea']
Teams = sorted(Teams)
Teams
```

```python
df = df[df['Team'].isin(Teams)].sort_values(by=['Team', 'Season'],ascending='True')
df
```

```python
teams_win = df.groupby('Team')['W'].sum()
teams_win
```

```python
teams_draw = df.groupby('Team')['D'].sum()
teams_loss = df.groupby('Team')['L'].sum()
```

```python
plt.close()
```

```python
fig, ax = plt.subplots()
ax.bar(Teams, teams_win, label='win')
ax.bar(Teams, teams_draw, bottom=teams_win, label='draw')
ax.bar(Teams, teams_loss, bottom=teams_draw+teams_win, label='loss')
ax.set_xticklabels(Teams, rotation=90)
ax.set_ylabel('Number of wins')
plt.title('2010-21 EPL W/D/L')
plt.legend()
plt.show()
```

```python
df = pd.read_csv(PATH + 'EPL Standings 2010-2021.csv')
df
```

```python
winner = df[df['Pos'] == 1].sort_values(by=['Season'])
winner
```

```python
runner_up = df[df['Pos'] == 2].sort_values(by=['Season'])
runner_up
```

```python
fig, ax = plt.subplots()
ax.hist(winner['Pts'], label='Winner', bins=10, histtype='step')
ax.hist(runner_up['Pts'], label='Runner up', bins=10, histtype='step')
ax.set_xlabel('Pts')
ax.set_ylabel('Number of teams')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(winner['Pts'], label='Winner', bins=10, alpha=0.5)
ax.hist(runner_up['Pts'], label='Runner up', bins=10, alpha=0.5)
ax.set_xlabel('Pts')
ax.set_ylabel('Number of teams')
plt.legend()
plt.show()
```

```python
df_copy = df.copy()
df = df.sort_values(by=['Season', 'Pts'], ascending=[True, False])
df
```

```python
gf_mean = df.groupby('Season')['GF'].mean()
pts_mean = df.groupby('Season')['Pts'].mean()
df.groupby('Season')['GF'].mean()
```

```python
gf_std = df.groupby('Season')['GF'].std()
pts_std = df.groupby('Season')['Pts'].std()
df.groupby('Season')['GF'].std()
```

```python
winner.std(axis=0)
```

```python
fig, ax = plt.subplots()

ax.errorbar(winner['Season'], gf_mean, yerr=gf_std/50, label='GF')
ax.errorbar(runner_up['Season'], pts_mean, yerr=pts_std/50, label='PTS')
ax.set_xticklabels(winner['Season'], rotation=90)
plt.legend()
plt.title('Seasonal mean GF & PTS of every teams')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot([winner['Pts'],runner_up['Pts']])
ax.set_xticklabels(['Winner','Runner-up'])
ax.set_ylabel('Pts')
plt.show()
```

```python
teams = list(df.groupby('Team').groups.keys())
mean_pts = df.groupby('Team')['Pts'].mean()
std_pts = df.groupby('Team')['Pts'].std()
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()
for team, mean, std in zip(teams, mean_pts, std_pts):
    ax.bar(team, mean, yerr=std)
ax.set_ylabel('Pts')
ax.set_xticklabels(teams, rotation=90)
plt.title('EPL 10-21 average pts of all teams')
plt.show()
```
